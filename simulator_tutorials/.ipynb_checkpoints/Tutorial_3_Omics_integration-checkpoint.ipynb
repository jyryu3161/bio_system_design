```python
import LAD
import pandas as pd
import copy
from cobra.io import read_sbml_model
import GPR_manipulation
import omics_score_calculation

def calculate_reaction_score(cobra_model, expression_score):
    expression_socre_string = {}
    for gene_id in expression_score:
        expression_socre_string[str(gene_id)] = expression_score[gene_id]
    reaction_weights = {}
    for each_reaction in cobra_model.reactions:        
        temp_expression_socre_string = copy.deepcopy(expression_socre_string)
        compartments = [metabolite.compartment for metabolite in each_reaction.reactants + each_reaction.products]
        compartments = list(set(compartments))
        reaction_weights[each_reaction.id] = 0.0
        GPR_association = each_reaction.gene_reaction_rule
        genes = [gene.id for gene in each_reaction.genes]
        genes = list(set(genes) & set(temp_expression_socre_string.keys()))
        if len(genes) > 1:
            GPR_list = GPR_manipulation.convert_string_GPR_to_list_GPR(GPR_association)
            expression = omics_score_calculation.GPR_score_calculation(GPR_list, temp_expression_socre_string)
            reaction_weights[each_reaction.id] = expression
        elif len(genes) == 1:
            gene_id = genes[0].strip()
            if gene_id in temp_expression_socre_string:
                expression = temp_expression_socre_string[gene_id]
                reaction_weights[each_reaction.id] = expression
    return reaction_weights

# Model 및 데이터 로드
model_file = './iJR904.xml'
cobra_model = read_sbml_model(model_file)
biomass_reaction = "BIOMASS_Ecoli"
df = pd.read_csv('./my_rna_seq.csv', index_col=0)

# Simulator 초기화
sim = LAD.LAD()
sim.load_cobra_model(cobra_model)

# 초기 FBA로 최대 성장률 계산
_, _, flux_results = sim.run_FBA(new_objective=biomass_reaction)
max_growth_rate = flux_results[biomass_reaction] * 0.9

# 모든 반응의 ID 가져오기
all_reactions = [rxn.id for rxn in cobra_model.reactions]

# 결과 저장용 DataFrame 초기화
flux_no_constraint = pd.DataFrame(index=all_reactions)
flux_with_constraint = pd.DataFrame(index=all_reactions)

# 각 샘플에 대해 flux 계산
for i, each_col in enumerate(df.columns):
    print(f"\nProcessing sample {i+1}/{len(df.columns)}: {each_col}")
    
    expression_score = df[each_col].to_dict()
    reaction_weights = calculate_reaction_score(cobra_model, expression_score)
    
    # 1. Growth rate 제약 없이 발현 데이터에 fitting
    status, obj_val, flux_results_no_const = sim.run_LP_fitting(
        opt_flux=reaction_weights, 
        flux_constraints={}
    )
    
    # 결과를 DataFrame에 저장
    flux_no_constraint[each_col] = pd.Series(flux_results_no_const)
    
    # 2. Growth rate 제약을 주고 발현 데이터에 fitting
    flux_constraints = {}
    flux_constraints[biomass_reaction] = [max_growth_rate*0.9, max_growth_rate]
    status, obj_val, flux_results_with_const = sim.run_LP_fitting(
        opt_flux=reaction_weights, 
        flux_constraints=flux_constraints
    )
    
    # 결과를 DataFrame에 저장
    flux_with_constraint[each_col] = pd.Series(flux_results_with_const)
    
    print(f"  - Biomass without constraint: {flux_results_no_const.get(biomass_reaction, 0):.4f}")
    print(f"  - Biomass with constraint: {flux_results_with_const.get(biomass_reaction, 0):.4f}")

# NaN을 0으로 채우기 (반응이 없는 경우)
flux_no_constraint = flux_no_constraint.fillna(0)
flux_with_constraint = flux_with_constraint.fillna(0)

# CSV 파일로 저장
flux_no_constraint.to_csv('./flux_distribution_no_constraint.csv')
flux_with_constraint.to_csv('./flux_distribution_with_constraint.csv')

print("\n=== 저장 완료 ===")
print(f"No constraint: flux_distribution_no_constraint.csv")
print(f"With constraint: flux_distribution_with_constraint.csv")
```
