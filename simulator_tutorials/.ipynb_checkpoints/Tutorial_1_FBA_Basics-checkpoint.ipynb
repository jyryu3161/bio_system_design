## 5. Theoretical yield 계산

- FBA(Flux Balance Analysis)에서 Theoretical yield(이론적 수율)은
주어진 기질(substrate)로부터 생성 가능한 최대 생성물(product)의 양을 의미
- 이는 대사 네트워크가 가질 수 있는 최대 효율적인 생산 한계치를 이론적으로 계산한 값으로,
실험적 조건(효소 속도, 조절, 비균질 환경 등)을 고려하지 않은 최적화된 이상 상태의 수율

```python
# new target reaction: EX_lac__L_e
# biomass reaction: BIOMASS_Ec_iML1515_core_75p37M
# COBRApy에서 모델 로드
model = cobra.io.load_model("iML1515")
model.reactions.ATPM.lower_bound = 0.0
model.reactions.ATPM.upper_bound = 1000.0

sim = Simulator()
sim.load_cobra_model(model)

target_reaction = "EX_lac__L_e"
new_flux_constraints = {}
new_flux_constraints["EX_glc__D_e"] = [-1.0, 0.0]
status, objective_value, fluxes = sim.run_FBA(new_objective=target_reaction, flux_constraints=new_flux_constraints, 
                                          internal_flux_minimization=False, mode='max')

print(status, objective_value)
print(fluxes["BIOMASS_Ec_iML1515_core_75p37M"])
print(fluxes["EX_lac__L_e"])
print(fluxes["EX_glc__D_e"])

```

## 6. 주요 교환 반응 분석


```python
model = cobra.io.load_model("iML1515")
sim = Simulator()
sim.load_cobra_model(model)

biomass_reaction = "BIOMASS_Ec_iML1515_core_75p37M"
status, objective_value, fluxes = sim.run_FBA(new_objective=biomass_reaction, internal_flux_minimization=True)

# 주요 교환 반응
exchange_reactions = {
    'EX_glc__D_e': 'Glucose 흡수',
    'EX_o2_e': 'Oxygen 흡수',
    'EX_co2_e': 'CO2 분비',
    'EX_ac_e': 'Acetate 분비',
    'EX_etoh_e': 'Ethanol 분비',
    'EX_for_e': 'Formate 분비'
}

print("=== 주요 교환 반응 ===")
exchange_data = []
for rxn_id, name in exchange_reactions.items():
    if rxn_id in fluxes:
        flux = fluxes[rxn_id]
        print(f"{name:20s}: {flux:8.4f} mmol/gDW/hr")
        exchange_data.append({'반응': name, '플럭스': flux})

# DataFrame으로 변환
df_exchange = pd.DataFrame(exchange_data)
df_exchange
```

## 8. MOMA 분석: Constraint와 Knockout

MOMA (Minimization of Metabolic Adjustment)는 유전자 knockout이나 환경 변화 후 대사 네트워크가 어떻게 적응하는지 예측합니다.


```python
model = cobra.io.load_model("e_coli_core")
sim = Simulator()
sim.load_cobra_model(model)

biomass_reaction = "BIOMASS_Ecoli_core_w_GAM"
status, objective_value, wt_fluxes = sim.run_FBA(new_objective=biomass_reaction, internal_flux_minimization=True)

wt_growth = wt_fluxes[biomass_reaction]
wt_succ = wt_fluxes.get('EX_succ_e', 0)
print(f"야생형 성장: {wt_growth:.4f} hr⁻¹")
print(f"야생형 숙신산 생산: {wt_succ:.4f} mmol/gDW/hr")

# MOMA 분석 1: PGI knockout
print("\n=== MOMA 분석 1: PGI 반응 knockout ===")
knockout_constraints = {
    'PGI': (0, 0)
}

status_moma, distance, moma_fluxes = sim.run_MOMA(
    wild_flux=wt_fluxes,
    flux_constraints=knockout_constraints
)

if status_moma == 'optimal':
    moma_growth = moma_fluxes[sim.objective]
    moma_succ = moma_fluxes.get('EX_succ_e', 0)
    
    print(f"MOMA 예측 성장: {moma_growth:.4f} hr⁻¹ ({(1-moma_growth/wt_growth)*100:.1f}% 감소)")
    print(f"MOMA 예측 숙신산 생산: {moma_succ:.4f} mmol/gDW/hr")
    print(f"대사 거리 (L1 norm): {distance:.4f}")
    
    # 가장 많이 변화된 반응들
    flux_changes = {
        rxn: abs(moma_fluxes[rxn] - wt_fluxes[rxn])
        for rxn in wt_fluxes.keys()
    }
    top_5_changes = sorted(flux_changes.items(), key=lambda x: x[1], reverse=True)[:5]
    
    print("\n가장 많이 변화된 반응 (Top 5):")
    for rxn, change in top_5_changes:
        if change > 1e-6:
            print(f"  {rxn}: {wt_fluxes[rxn]:.3f} → {moma_fluxes[rxn]:.3f} (Δ={change:.3f})")
else:
    print("MOMA 최적화 실패!")
```

## 9. Gene knock-out target prediction simulation

```python
model = cobra.io.load_model("e_coli_core")
sim = Simulator()
sim.load_cobra_model(model)

biomass_reaction = "BIOMASS_Ecoli_core_w_GAM"
target_reaction = "EX_succ_e"

status, objective_value, wt_fluxes = sim.run_FBA(new_objective=biomass_reaction, internal_flux_minimization=True)

fp = open('./simulation_results.txt', 'w')
fp.write('%s\t%s\t%s\t%s\n'%("Gene", "Reactions", "Growth", "Production rate"))
print('%s\t%s\t%s\t%s'%("Gene", "Reactions", "Growth", "Production rate"))
for each_gene in model.genes:
    ko_reaction_constraints = {}
    ko_reactions = []
    for each_reaction in each_gene.reactions:
        ko_reaction_constraints[each_reaction.id] = [0.0, 0.0]
        ko_reactions.append(each_reaction.id)
        
    status_moma, distance, moma_fluxes = sim.run_MOMA(
        wild_flux=wt_fluxes,
        flux_constraints=ko_reaction_constraints
    )
    
    if status_moma == "optimal":
        growth = moma_fluxes[biomass_reaction]
        target_production = moma_fluxes[target_reaction]
        ko_reaction_str = ';'.join(ko_reactions)
        fp.write('%s\t%s\t%s\t%s\n'%(each_gene, ko_reaction_str, growth, target_production))
        print('%s\t%s\t%s\t%s'%(each_gene, ko_reaction_str, growth, target_production))

fp.close()
```

```python
model = cobra.io.load_model("e_coli_core")
sim = Simulator()
sim.load_cobra_model(model)

biomass_reaction = "BIOMASS_Ecoli_core_w_GAM"
target_reaction = "EX_succ_e"

# 먼저 최대 생산량을 계산하고, 일부분을 constraint로 걸어준 조건을 WT flux distribution으로 사용
status, objective_value, fluxes = sim.run_FBA(new_objective=target_reaction)
max_target_production = fluxes[target_reaction]

new_constraints = {}
new_constraints[target_reaction] = [max_target_production*0.01, 1000]
status, objective_value, wt_fluxes = sim.run_FBA(new_objective=biomass_reaction, flux_constraints = new_constraints, internal_flux_minimization=True)

fp = open('./simulation_results_v2.txt', 'w')
fp.write('%s\t%s\t%s\t%s\n'%("Gene", "Reactions", "Growth", "Production rate"))
print('%s\t%s\t%s\t%s'%("Gene", "Reactions", "Growth", "Production rate"))
for each_gene in model.genes:
    ko_reaction_constraints = {}
    ko_reactions = []
    for each_reaction in each_gene.reactions:
        ko_reaction_constraints[each_reaction.id] = [0.0, 0.0]
        ko_reactions.append(each_reaction.id)
        
    status_moma, distance, moma_fluxes = sim.run_MOMA(
        wild_flux=wt_fluxes,
        flux_constraints=ko_reaction_constraints
    )
    
    if status_moma == "optimal":
        growth = moma_fluxes[biomass_reaction]
        target_production = moma_fluxes[target_reaction]
        ko_reaction_str = ';'.join(ko_reactions)
        fp.write('%s\t%s\t%s\t%s\n'%(each_gene, ko_reaction_str, growth, target_production))
        print('%s\t%s\t%s\t%s'%(each_gene, ko_reaction_str, growth, target_production))

fp.close()
```

# Production of non-native target metabolite

- PHB는 미생물이 탄소와 에너지를 저장하기 위해 합성하는 생분해성 고분자(폴리에스터)
- 화학식은 (C₄H₆O₂)ₙ, 즉 3-hydroxybutyrate 단위가 중합된 형태입
- 이는 석유 기반 플라스틱(PET, PP 등)을 대체할 수 있는 친환경 플라스틱으로 주목받고 있으며, E. coli, Ralstonia eutropha, Cupriavidus necator 등 다양한 미생물이 PHB를 생산할 수 있음

## 모델 수정

```python
from cobra import Model, Reaction, Metabolite
import cobra
from cobra.io import write_sbml_model, save_json_model

# COBRApy ≥0.22 가정
model = cobra.io.load_model("e_coli_core")

# 2) 새 대사물질 정의 (cytosol, _c)
#    - core 모델에 없는 중간체와 PHB 폴리머
aacoa_c   = Metabolite("aacoa_c",  name="Acetoacetyl-CoA", compartment="c")
hbcoa_c   = Metabolite("3hbcoa_c", name="(R)-3-Hydroxybutyryl-CoA", compartment="c")
phb_c     = Metabolite("phb_c",    name="Polyhydroxybutyrate (PHB) [polymer unit]", compartment="c")


# 3) phaA 반응: 2 Acetyl-CoA -> Acetoacetyl-CoA + CoA
#    (EC: β-ketothiolase; reversible로도 쓰이지만 생산 방향으로 설정)
phaA = Reaction("PHA_A")
phaA.name = "β-ketothiolase (phaA)"
phaA.lower_bound = 0.0       # forward only
phaA.upper_bound = 1000.0
phaA.add_metabolites({
    model.metabolites.accoa_c: -2.0,   # 2 Acetyl-CoA
    aacoa_c: 1.0,                      # Acetoacetyl-CoA
    model.metabolites.coa_c: 1.0       # CoA-SH
})

# 4) phaB 반응: Acetoacetyl-CoA + NADPH + H+ -> 3HB-CoA + NADP+

phaB = Reaction("PHA_B")
phaB.name = "acetoacetyl-CoA reductase (phaB)"
phaB.lower_bound = 0.0
phaB.upper_bound = 1000.0

phaB.add_metabolites({
    aacoa_c: -1.0,
    model.metabolites.nadph_c: -1.0,
    model.metabolites.h_c: -1.0,
    hbcoa_c: 1.0,
    model.metabolites.nadp_c: 1.0
})


phaC = Reaction("PHA_C")
phaC.name = "PHB synthase (phaC)"
phaC.lower_bound = 0.0
phaC.upper_bound = 1000.0
phaC.add_metabolites({
    hbcoa_c: -1.0,
    phb_c: 1.0,
    model.metabolites.coa_c: 1.0
})

# 6) PHB demand 반응 (축적 용도)
#    호환성 위해 수동으로 추가
dm_phb = Reaction("EX_phb_c")
dm_phb.name = "Exchange for PHB (polymer)"
dm_phb.lower_bound = 0.0
dm_phb.upper_bound = 1000.0
dm_phb.add_metabolites({phb_c: -1.0})

# 7) 모델에 반응 추가
model.add_reactions([phaA, phaB, phaC, dm_phb])

EX_GLC = model.reactions.get_by_id("EX_glc__D_e")
EX_O2  = model.reactions.get_by_id("EX_o2_e")
EX_NH4 = model.reactions.get_by_id("EX_nh4_e")

# 기본 탄소 공급
EX_GLC.lower_bound = -10.0   # mmol/gDW/h
# 산소는 기본 자유 (필요 시 무산소로: EX_O2.lower_bound = 0.0)
# 질소 제한: 더 강하게 제한할수록 PHB 축적 유리
EX_NH4.lower_bound = -1.0    # 예: -1 mmol/gDW/h 로 제한

# 9) 목적함수를 PHB demand로 설정하고 최적화
model.objective = dm_phb
solution = model.optimize()

print("=== PHB FBA 결과 ===")
print(f"Objective (PHB rate)    : {solution.objective_value:.6f} mmol/gDW/h")
print(f"Glucose uptake          : {-solution.fluxes['EX_glc__D_e']:.6f} mmol/gDW/h (positive = consumed)")
print(f"O2 uptake               : {-solution.fluxes['EX_o2_e']:.6f} mmol/gDW/h")
print(f"NH4 uptake              : {-solution.fluxes['EX_nh4_e']:.6f} mmol/gDW/h")
print(f"Biomass flux            : {solution.fluxes['BIOMASS_Ecoli_core_w_GAM']:.6f} 1/h")

# 10) 수율 계산 (PHB 단량체 기준 mol/mol glucose)
glc_uptake = -solution.fluxes["EX_glc__D_e"]  # 양수로 변환
phb_prod   = solution.fluxes["EX_phb_c"]
yield_phb  = phb_prod / glc_uptake if glc_uptake > 1e-9 else float("nan")
print(f"Theoretical PHB yield   : {yield_phb:.6f} mol PHB-monomer / mol glucose")

# 11) 확인용: 추가한 반응들의 플럭스 출력
for rxn_id in ["PHA_A", "PHA_B", "PHA_C", "EX_phb_c"]:
    print(f"{rxn_id:>10s} : {solution.fluxes[rxn_id]:.6f}")

write_sbml_model(model, "ecoli_core_phb.xml")
```
